- 항등함수
    - 항등함수는 입력과 출력이 동일한 결과 즉 입력하는 값을 그대로 출력하는 값으로 처리하는 함수입니다.

```python
def identity(x):
    return x
```

```python
x = np.arange(-10,11)
y = identity(x)
x == y
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(x,y)
# spines()
plt.show()
```

- 계단함수
    - 계단함수는 입력값의 결과가 0과 1로 표시합니다. 직선으로 표시가 되지 않는 함수라 비선형 함
    수입니다. 함수의 수식을 보면 입력값 x 가 0이하이면 출력값 y는 0이고, x가 0보다 크면 y는 1이
    되는 함수입니다. 계단함수의 단점은 0과 1 사이의 값을 표시할 수 없습니다. 이 함수는 주로 퍼셉
    트론을 처리할 때 많이 사용됩니다.

```python
x = np.arange(-5,5,0.1)
```

```python
def step(x):
    return np.where (x>0, 1,0)
```

```python
y = step(x)
```

```python
y
```

```python
plt.plot(x,y)
plt.grid()
# spines()
plt.show()
```

- 하이퍼볼릭 탄젠트 함수
    - 시그모이드 함수처럼 자연상수 e를 제곱해서 분모는 합하고 분자는 빼는 구조입니다. 
    계산을 하면 -1과 1 사이의 값을 0을 중심으로 대칭형태로 값을 보여줍니다.

```python
x = np.arange(-5,5,0.1)
```

```python
y = np.tanh(x)
```

```python
plt.plot(x,y)
plt.grid()
# spines()
plt.show()
```

- 시그모이드 함수
    - 시그모이드 함수는 자연상수를 지수로 사용해서 작은 값과 큰 값을 특정 범위 내의 값으로 처리합니다. 그래서 입력값이 어떤 값이 주어져도 출력값은 0과 1 사이로 반환합니다. 

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5,5,0.1)
y = sigmoid(x)
```

```python
x.shape, y.shape
```

```python
plt.plot(x,y)
plt.grid()
# spines()
plt.show()
```

- 렐루 함수
    - 출력값은 노드의 수가 n이라면 이 노드의 전체를 자연상수 e의 제곱의 합을 구하고 특정
    노드는 자연상수 e의 제곱으로 구해서 나눈 값을 반환합니다. 이 함수의 값은 0보다 크
    고 1보다 작은 값을 반환합니다. 이 함수는 분류를 구하는 출력층에 주로 사용합니다.

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = np.arange(-5,5,0.1)
y = relu(x)
```

```python
plt.plot(x,y)
plt.grid()
# spines()
plt.show()
```

- 소프트맥스 함수
    - 출력값은 노드의 수가 n이라면 이 노드의 전체를 자연상수 e의 제곱의 합을 구하고 특정
    노드는 자연상수 e의 제곱으로 구해서 나눈 값을 반환합니다. 이 함수의 값은 0보다 크
    고 1보다 작은 값을 반환합니다. 이 함수는 분류를 구하는 출력층에 주로 사용합니다.

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
x = np.arange(1,5)
y = softmax(x)
```

```python
y
```

```python
np.sum(y)
```

```python
ratio = y
labels = y
plt.pie(ratio, labels=labels, shadow=True, startangle=90)
plt.show()
```
