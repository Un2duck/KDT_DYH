```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
pop = pd.read_excel('population.xlsx')
pop
```

```python
pop.drop(columns='상품ID', inplace=True)
pop.head()
```

```python
gender_age_DF = pop.iloc[:,:6]
gender_age_DF
```

```python
gender_age_group = gender_age_DF.groupby(['상품명','성별','나이'])
```

```python
product_sum_DF = gender_age_group.value_counts().to_frame()
product_sum_DF
```

```python
gender_age_DF
```

```python
def check_gender_age(gender, age):
    indexing = gender_age_DF[(gender_age_DF['성별'] == gender) & (gender_age_DF['나이'] == age)]
    return indexing
```

```python
check_gender_age('여성', '36~40')
```

```python
check_gender_age('여성', '36~40').sort_values(by='결제금액', ascending=False).head(5)
```

```python
check_gender_age('남성', '36~40').sort_values(by='결제금액', ascending=False).head(5)
```

```python
# 나이 구간 설정 (연령대 구간은 '10대'부터 '70대 이상'까지)
bins = [10, 20, 30, 40, 50, 60, 70, 100]

# 각 구간에 해당하는 레이블 지정
labels = ['10대', '20대', '30대', '40대', '50대', '60대', '70대 이상']

# '나이' 컬럼을 연령대로 변환
gender_age_DF['연령대'] = pd.cut(gender_age_DF['나이'].str.extract('(\d+)', expand=False).astype(int), bins=bins, labels=labels, right=False)
gender_age_DF
```

```python
def check_gender_age2(gender, age):
    indexing = gender_age_DF[(gender_age_DF['성별'] == gender) & (gender_age_DF['연령대'] == age)]
    return indexing
```

```python
product1 = check_gender_age2('여성', '30대').sort_values(by='결제금액', ascending=False).head(5)
product1
```

```python
plt.barh(product1['상품명'], product1['결제금액'])
plt.legend()
plt.gca().invert_yaxis()
plt.show()
```

```python
age_list = ['10대','20대','30대','40대','50대','60대','70대 이상']
woman_product_list = []
for i in range(7):
    woman_product_list.append(check_gender_age2('여성', age_list[i]).sort_values(by='결제금액', ascending=False).head(5))

woman_product_list
```

```python
for i in range(7):
    plt.barh(woman_product_list[i]['상품명'], woman_product_list[i]['결제금액'], color='lightcoral', label='누적 결제금액')
    plt.title(f'여성 {age_list[i]}의 인기 상품', size=25)
    plt.gca().invert_yaxis()
    plt.legend()
    # plt.savefig(f'woman_popular{i}.png', transparent=True)  # 배경을 투명하게 저장
    plt.show()
```

```python
age_list = ['10대','20대','30대','40대','50대','60대','70대 이상']
man_product_list = []
for i in range(7):
    man_product_list.append(check_gender_age2('남성', age_list[i]).sort_values(by='결제금액', ascending=False).head(5))

man_product_list
```

```python
for i in range(7):
    plt.barh(man_product_list[i]['상품명'], man_product_list[i]['결제금액'], color='cornflowerblue', label='누적 결제금액')
    plt.title(f'남성 {age_list[i]}의 인기 상품', size=25)
    plt.gca().invert_yaxis()
    plt.legend()
    # plt.savefig(f'man_popular{i}.png', transparent=True)  # 배경을 투명하게 저장
    plt.show()
```
