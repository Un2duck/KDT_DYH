```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# 1. 데이터 불러오기
df = pd.read_excel('population.xlsx')

# 2. 피처 선택 (예: 결제금액, 결제상품수량)
features = df[['결제금액', '결제상품수량']]

# 3. 데이터 스케일링
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# 4. K-Means 군집화
kmeans = KMeans(n_clusters=3, random_state=42)  # 3개의 클러스터로 분할
df['Cluster'] = kmeans.fit_predict(scaled_features)

# 5. PCA로 2D 시각화
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

plt.figure(figsize=(10, 8))
plt.scatter(pca_features[:, 0], pca_features[:, 1], c=df['Cluster'], cmap='viridis', s=50)
plt.title('K-Means Clustering of Population Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. 데이터 스케일링
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# 2. DBSCAN 군집화
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['Cluster'] = dbscan.fit_predict(scaled_features)

# 3. PCA로 2D 시각화
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

plt.figure(figsize=(10, 8))
plt.scatter(pca_features[:, 0], pca_features[:, 1], c=df['Cluster'], cmap='viridis', s=50)
plt.title('DBSCAN Clustering of Population Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# 1. 데이터 스케일링
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# 2. Agglomerative Clustering
agglo = AgglomerativeClustering(n_clusters=3)
df['Cluster'] = agglo.fit_predict(scaled_features)

# 3. PCA로 2D 시각화
pca = PCA(n_components=2)
pca_features = pca.fit_transform(scaled_features)

plt.figure(figsize=(10, 8))
plt.scatter(pca_features[:, 0], pca_features[:, 1], c=df['Cluster'], cmap='viridis', s=50)
plt.title('Agglomerative Clustering of Population Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```
