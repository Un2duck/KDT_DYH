```python
import pandas as pd

# file = '따끈따끈_수도사용량_테스트데이터_12.06.(수정).xlsx'
# water_sample = pd.read_excel(file)
# water_sample.head()

file = '따끈따끈_전기사용량_테스트데이터(12.9.).xlsx'
electric_sample = pd.read_excel(file)
electric_sample.head()
```

```python
# Define office_id and managed_entity_id for all rows
office_id = "OFF001"
managed_entity_id = "ME001"

# Melt the table to normalize data
# normalized_data = water_sample.melt(id_vars=["member_id"], var_name="datetime", value_name="water_usage")

normalized_data = electric_sample.melt(id_vars=["member_id"], var_name="datetime", value_name="electric_usage")


# # Extract electric_date and electrict to the correct format
# normalized_data["water_date"] = normalized_data["datetime"].str.extract(r"(\d{4}-\d{2}-\d{2})")
# normalized_data["office_id"] = office_id
# normalized_data["managed_entity_id"] = managed_entity_id


normalized_data["electric_date"] = normalized_data["datetime"].str.extract(r"(\d{4}-\d{2}-\d{2})")
normalized_data["office_id"] = office_id
normalized_data["managed_entity_id"] = managed_entity_id

# Rearrange columns to match water table structure
# water_table_data = normalized_data[["office_id", "member_id", "managed_entity_id", "water_date", "water_usage"]]

electric_table_data = normalized_data[["office_id", "member_id", "managed_entity_id", "electric_date", "electric_usage"]]
```

```python
# Save to a CSV file for database insertion
# water_file_path = "all_converted_water_sample(1206).csv"
# water_table_data.to_csv(water_file_path, index=False)

# water_file_path

electric_file_path = "all_converted_electric_sample(1209).csv"
electric_table_data.to_csv(electric_file_path, index=False)

electric_file_path
```
