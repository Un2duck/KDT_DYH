### 전체 생존율, 사망율 계산하기

```python
# 생존인원수
survived_num=titan['survived'].sum()
print(f'전체 생존 인원수: {survived_num}')

# 전체 탑승인원수

total_num=len(titan.index)
print(f'전체 탑승 인원수: {total_num}')

# 생존율
print(f'생존율: {round(survived_num/total_num, 2)}%')
```

### 탑승지 (S,C,Q) 별 생존율 계산하기

#### S (Southampton) 생존율

```python
# S 탑승지 Series
S_SR=titan[titan['embarked'] == 'S']

# S(Southampton) 전체인원수
S_total=len(S_SR.index)
print(f'S 전체인원수: {S_total}')

# S(Southampton) 생존인원수
Survived_S=S_SR['survived'].sum()
print(f'S 생존인원수: {Survived_S}')

# S(Southampton) 생존율
print(f'S 생존율: {round(Survived_S/S_total, 2)}%')
```

#### C (Cherbourg) 생존율

```python
# C 탑승지 Series
C_SR=titan[titan['embarked'] == 'C']

# C(Cherbourg) 전체인원수
C_total=len(C_SR.index)
print(f'C 전체인원수: {C_total}')

# C(Cherbourg) 생존인원수
Survived_C=C_SR['survived'].sum()
print(f'C 생존인원수: {Survived_C}')

# C(Cherbourg) 생존율
print(f'C 생존율: {round(Survived_C/C_total, 2)}%')
```

#### Q (Queenstown) 생존율

```python
# Q 탑승지 Series
Q_SR=titan[titan['embarked'] == 'Q']

# Q (Queenstown) 전체인원수
Q_total=len(Q_SR.index)
print(f'Q 전체인원수: {Q_total}')

# Q (Queenstown) 생존인원수
Survived_Q=Q_SR['survived'].sum()
print(f'Q 생존인원수: {Survived_Q}')

# Q (Queenstown) 생존율
print(f'Q 생존율: {round(Survived_Q/Q_total, 2)}%')
```

### Embarked 의 결측치 2개 값 예상하여 생존율 재 계산하기

```python
titan[['embarked','survived','alone']]
```

```python
# Embarked의 결측치 2개 찾기

no_embarked=titan[(titan['embarked'] != 'Q') & (titan['embarked'] != 'S') & (titan['embarked'] != 'C') == True]
no_embarked
```

```python
# 위의 2명은 생존했기 때문에 아래의 S,C,Q 생존률 확률에 더하여 재 계산할 수 있음.
# 기존 889명은 아래와 같은 확률이었음.

print(f'기존 S 생존율: {round(Survived_S/S_total, 2)}%') # 217명 / 644명
print(f'기존 C 생존율: {round(Survived_C/C_total, 2)}%') # 93명 / 168명  
print(f'기존 Q 생존율: {round(Survived_Q/Q_total, 2)}%') # 30명 / 77명
print()

# 38세 여성을 A, 62세 여성을 B로 명명하여,
# 경우의 수는 모두 6가지임.
# (S,S), (C,C), (Q,Q), (S,C), (C,Q), (S,Q)
# 하지만!
```

```python
# A와 B 모두 *alone* 이고, fare 값이 *80*이므로
no_embarked[['alone', 'fare']]
```

```python
# 혼자탄 사람들의 fare 값을 기준으로 볼 때
# Q_fare 최대값이 24.15이므로
# 두 사람은 'S'나 'C'일 수밖에 없을 것으로 예측됨.

print('혼자탄 S의 fare 최대값', alone_fare_S['fare'].max(), end='\n\n')
print('혼자탄 C의 fare 최대값', alone_fare_C['fare'].max(), end='\n\n')
print('혼자탄 Q의 fare 최대값', alone_fare_Q['fare'].max(), end='\n\n')
```

```python
# 따라서 경우의 수는 아래 3가지 중 하나.

# (1)
# 모두 S인 경우,
print('(case1) 모두 S인 경우')
print(f'(case1) S 생존율: {round((Survived_S+2)/(S_total+2), 2)}%') # 217명+2명 / 644명+2명
print(f'(case1) C 생존율: {round(Survived_C/C_total, 2)}%') # 93명 / 168명  
print(f'(case1) Q 생존율: {round(Survived_Q/Q_total, 2)}%') # 30명 / 77명
print()

# (2)
# 모두 C인 경우,
print('(case2) 모두 C인 경우')
print(f'(case2) S 생존율: {round(Survived_S/S_total, 2)}%') # 217명 / 644명
print(f'(case2) C 생존율: {round((Survived_C+2)/(C_total+2), 2)}%') # 93명+2명 / 168명+2명
print(f'(case2) Q 생존율: {round(Survived_Q/Q_total, 2)}%') # 30명 / 77명
print()

# (3)
# S,C인 경우,
print('(case4) 각각 S,C인 경우')
print(f'(case4) S 생존율: {round((Survived_S+1)/(S_total+1), 2)}%') # 217명+1명 / 644명+1명
print(f'(case4) C 생존율: {round((Survived_C+1)/(C_total+1), 2)}%') # 93명+1명 / 168명+1명
print(f'(case4) Q 생존율: {round(Survived_Q/Q_total, 2)}%') # 30명 / 77명
print()
```

### (Bonus) 좌석요금을 내지 않은 사람들 생존율 계산하기

```python
# fare == 0 인 사람들
nofare=titan[titan['fare'] == 0]
```

```python
nofare.info()
nofare
```

```python
nofare['embarked'].value_counts()
```

```python
nofare['who'].value_counts()
```

```python
# 좌석요금을 내지 않은 사람은 모두 남자이다.
# 좌석요금을 내지 않은 사람의 탑승지는 모두 S (Southampton)이다.
# 좌석요금을 내지 않은 사람은 모두 혼자 탑승했다.
# 좌석요금을 내지 않은 사람 중 생존자는 단 한명이다.
```

```python
nofare.describe()
```

```python
# (좌석요금을 내지 않는 사람들) 생존자 수
nofare_survived=nofare['survived'].sum()
print(f'(좌석요금을 내지 않은 사람들) 생존자 수: {nofare_survived}')

# (좌석요금을 내지 않는 사람들) 전체 인원수
nofare_total=len(nofare.index)
print(f'(좌석요금을 내지 않은 사람들) 전체 인원수: {nofare_total}')

# (좌석요금을 내지 않는 사람들) 생존율
print(f'(좌석요금을 내지 않은 사람들) 생존율: {round(nofare_survived/nofare_total, 2)}%')
```
