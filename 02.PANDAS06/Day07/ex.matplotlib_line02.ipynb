```python
# - figure 객체 생성
# - 생성된 figure 객체 안에 subplot 객체 생성
# - 그래프를 표시 방식 지정 : 가로/세로 몇개 그래프 넣을지 지정
# - subplot에 그래프를 넣기
```

```python
# -> 하나의 Figure에 수력/화력/원자력/합계 그래프 넣기
fig=plt.figure(figsize=[10, 10])

# - add_subplot(행수, 열수, 번호)
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

# - 각 subplot에 그래프 넣기
ax1.plot(dataDF.loc[dataDF.index[1]], 'bo--', label='남한(수력)')
ax2.plot(dataDF.loc[dataDF.index[2]], 'bo--', label='남한(화력)')
ax3.plot(dataDF.loc[dataDF.index[3]], 'bo--', label='남한(원자력)')
ax4.plot(dataDF.loc[dataDF.index[0]], 'bo--', label='남한(합계)')

# - 각 subplot별로 설정
ax1.set_title('남한 - 수력발전량')
ax1.set_xlabel('연도')
ax1.set_xticks(dataDF.columns)

ax2.set_title('남한 - 화력발전량')
ax2.set_xlabel('연도')

ax3.set_title('남한 - 원자력발전량')
ax3.set_xlabel('연도')

ax4.set_title('남한 - 발전량 합계')
ax4.set_xlabel('연도')

# - 화면에 띄우기
plt.tight_layout()
plt.show()
```

```python
# -> 하나의 Figure에 수력/화력/원자력/합계 그래프 넣기
fig=plt.figure()

# -> subplots 객체 생성: subplots() => axes 객체를 배열에 담아서 반환
#                                  => 행단위로 axes 담아서 반환

axes = fig.subplots(1,4)

# - subplot에 그래프 넣기
for i in range(4):
    axes[i].plot(dataDF.loc[dataDF.index[i]], 'bo--', label=dataDF.index.values[i])

# - 화면에 띄우기
plt.tight_layout()
plt.show()
```
