### [10. Matplotlib 축 스케일 지정하기]

- X축 스케일 지정하기

```python
# [1] 모듈 생성
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = x ** 3

plt.plot(x, y)
plt.xscale('symlog')

plt.show()
```

- Y축 스케일 지정하기

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('log')

plt.show()
```

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('linear')

plt.show()
```

### [11. Matplotlib 여러 곡선 그리기]

- 기본 사용

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)
plt.show()
```

### [12. Matplotlib 그리드 설정하기]

- 기본 사용하기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.grid(True)

plt.show()
```

- 축 지정하기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)
plt.grid(True, axis='y')

plt.show()
```

- 스타일 설정하기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='springgreen', marker='^', markersize=9)
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')

plt.show()
```
