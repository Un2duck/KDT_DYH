```python
# [1] 모듈 생성
import numpy as np
import matplotlib.pyplot as plt
```

### [16. Matplotlib 막대 그래프 그리기]

- 기본 사용

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)

plt.show()
```

- 색상 지정하기

```python
fig=plt.figure(figsize=[10,10])
axes=fig.subplots(2,2)

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2', '#e35f62']

cnt=0
for i in range(2):
    for j in range(2):
        axes[i,j].bar(x, values, color=colors[cnt])
        cnt+=1

plt.xticks(x, labels=years)

plt.show()
```

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.bar(x, values, color=colors)
plt.xticks(x, years)

plt.show()
```

- 막대 폭 지정하기

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, width=0.4)
# plt.bar(x, values, width=0.6)
# plt.bar(x, values, width=0.8)
# plt.bar(x, values, width=1.0)
plt.xticks(x, years)

plt.show()
```

- 스타일 꾸미기

```python
x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, align='edge', edgecolor='lightgray',
        linewidth=5, tick_label=years)

plt.show()
```

### [17. Matplotlib 수평 막대 그래프 그리기]

- 기본 사용

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values)
plt.yticks(y, years)

plt.show()
```

- 색상 지정하기

```python
fig=plt.figure(figsize=[10,10])
axes=fig.subplots(2,2)

y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

colors = ['y', 'dodgerblue', 'C2', '#e35f62']

cnt=0
for i in range(2):
    for j in range(2):
        axes[i,j].barh(y, values, color=colors[cnt])
        cnt+=1
        
plt.yticks(y, years)
plt.show()
```

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

# plt.barh(y, values, color='y')
plt.barh(y, values, color='dodgerblue')
# plt.barh(y, values, color='C2')
# plt.barh(y, values, color='#e35f62')
plt.yticks(y, years)

plt.show()
```

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.barh(y, values, color=colors)
plt.yticks(y, years)

plt.show()
```

- 막대 높이 지정하기

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values, height=0.4)
# plt.barh(y, values, height=0.6)
# plt.barh(y, values, height=0.8)
# plt.barh(y, values, height=1.0)
plt.yticks(y, years)

plt.show()
```

- 스타일 꾸미기

```python
y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values, align='edge', edgecolor='#eee',
         linewidth=5, tick_label=years)

plt.show()
```

### [18. Matplotlib 산점도 그리기]

- 기본 사용

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x, y)
plt.show()
```

- 색상과 크기 지정하기

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
plt.plot([1], [1], 'o', markersize=20, c='#FF5733')
plt.scatter([2], [1], s=20**2, c='#33FFCE')

plt.text(0.5, 1.05, 'plot(markersize=20)', fontdict={'size': 14})
plt.text(1.6, 1.05, 'scatter(s=20**2)', fontdict={'size': 14})
plt.axis([0.4, 2.6, 0.8, 1.2])
plt.show()
```

- 투명도와 컬러맵 설정하기

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```
