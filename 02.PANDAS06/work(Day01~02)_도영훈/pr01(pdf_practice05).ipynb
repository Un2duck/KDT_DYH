## 구조화 데이터
#### 데이터프레임(DataFrame)
- 원소 값 변경
    * DataFrame객체 (일부분 또는 원소 선택 = 새로운 값)

```python
# 모듈 로딩
import pandas as pd
```

```python
# DataFrame() 함수로 데이터프레임 변환. 변수 df에 저장
exam_data={'이름':['서준','우현','인아'],
           '수학':[90, 80, 70],
           '영어':[98, 89, 95],
           '음악':[85, 95, 100],
           '체육':[100, 90, 90]}
```

```python
df=pd.DataFrame(exam_data)
df
```

```python
# '이름' 열을 새로운 인덱스로 지정하고, df 객체에 변경사항 반영
df.set_index('이름', inplace=True)
df
```

```python
# 데이터프레임 df의 특정 원소를 변경하는 방법:'서준'의 '체육' 점수
df.iloc[0][3] = 80
df
```

```python
df.loc['서준']['체육']=90
df
```

```python
df.loc['서준','체육']=100
df
```

```python
# 데이터프레임 df의 원소 여러 개를 변경하는 방법: '서준'의 '음악', '체육' 점수
df.loc['서준',['음악','체육']]=50
df
```

```python
df.loc['서준', ['음악', '체육']] = 100, 50
df
```

- 행, 열 위치 변경
    * DataFrame객체.transpose()
    * DataFrame객체.T

```python
df = df.transpose()
df
```

```python
df=df.T
df
```

- 특정 열을 행 인덱스로 설정
    * DataFrame객체.set_index(['열이름'])
    * DataFrame객체.set_index('열이름')
    원본 데이터 유지, 변경된 객체 반환 => inplace=True
