#### 함수 추가 : delete_dash

```python
def delete_dash(number):
    new_excelDF = list(excelDF.iloc[number])
    for i in range(len(excelDF.iloc[number])):
        if new_excelDF[i] == '-':
            new_excelDF[i] = 0
    excelDF.iloc[number]=new_excelDF
```

```python
# 4번째줄과 8번째줄의 '-'값을 0으로 변경
delete_dash(4)
delete_dash(8)
excelDF
```

#### 함수 추가 : row_total

```python
def row_total(row):
    total=0
    # 0번째줄 : list(excelDF.iloc[0,2:]) / N번째줄 : list(excelDF.iloc[N,2:])
    for i in range(len(list(excelDF.iloc[row,2:]))):
        total += list(excelDF.iloc[row,2:])[i]
    return total
```

```python
# 합계 리스트 만들기
total_list=[]

for i in excelDF.index:
    total_list.append(row_total(i))
print(total_list)
```

```python
excelDF['합계']=total_list
excelDF
```
