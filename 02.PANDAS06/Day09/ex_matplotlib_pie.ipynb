- 컬럼의 데이터 비율 확인

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비
F_DATA=r"C:\Users\KDP-50\OneDrive\바탕 화면\PANDAS06\Day08\auto_mpg.csv"
```

```python
# [3] CSV => DataFrame으로 저장
dataDF=pd.read_csv(F_DATA)
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산국가에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(), 2)
```

```python
# 전체 개수
# (1) dataDF['origin'].value_counts().sum()
dataDF['origin'].count()
```

```python
originSR=dataDF['origin'].value_counts()
x=originSR.index.to_list()
h=originSR.values.tolist()
x, h
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - bar 그래프
bar_graph=plt.bar(x, h, color=['darkturquoise', 'aqua', 'seagreen'])

# 막대에 라벨 올리기
plt.bar_label(bar_graph, h, label_type='center')

# x축 눈금 설정
plt.xticks([1,2,3],['USA','EU','JP'], rotation=45, fontsize=20)
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['USA','EU','JP'], autopct='%.2f%%')
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프 (반시계 방향)
plt.pie(h, labels=['USA','EU','JP'], autopct='%.2f%%', counterclock=False, startangle=90)
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['USA','EU','JP'], autopct='%.2f%%', counterclock=False, startangle=90
        , explode=[0, 0, 0.5])
plt.show()
```

```python
# - 자동차 생산국가에 대한 비율 시각화 - pie 그래프
plt.pie(h, labels=['USA','EU','JP'], autopct='%.2f%%', counterclock=False, startangle=90,
        wedgeprops={'linewidth':5, 'width':.5, 'edgecolor':'w'})
plt.show()
```
