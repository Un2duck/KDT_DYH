```python
# [1] 모듈 로딩
import pandas as pd
```

```python
# 소비자 물가 지수 변수명 : cpi
# 차량 소비량 : used_car
```

```python
# [2] 데이터 가져오기
file_name=r'C:\Users\KDP-17\EX_PANDAS6\daya_05\소비자물가상승률_2013.2023.xlsx'
file_name2=r'C:\Users\KDP-17\EX_PANDAS6\daya_05\자동차 내수시장 동향.xlsx'
file_name3=r'C:\Users\KDP-17\EX_PANDAS6\daya_05\자동차 등록 대수.xlsx'
```

```python
# [3] DF로 저장
cpi=pd.read_excel(file_name)
cpi
```

```python
# [3] DF로 저장
car_kor=pd.read_excel(file_name2,thousands=',')
car_kor
```

```python
# [3] DF로 저장
car_update=pd.read_excel(file_name3,thousands=',')
car_update
```

- 물가의 변동과 국산차 내수시장의 관계를 분석하기 위하기 때문에 데이터 간 연도를 맞추기 위해 2013년,2023년 데이터는 없앰
- 구분을 확실히 하기 위해 1번 열의 이름을 '목록'으로 설정

```python
# [4]
# [4-1] 소비자물가 지수 및 소비자 물가 상승률 데이터 확인 및 가공
cpi.info()
```

```python
# [4-2] cpi의 컬럼 이름 변경 및 필요한 행 및 열만 추출
cpi2=cpi.rename(columns={'Unnamed: 0':'목록'})
cpi2
```

```python
cpi3=cpi2.drop(['2013','2023'],axis=1)
```

```python
cpi3=cpi3[:2]
```

```python
cpi3
```

```python
# [5-1] 자동차 내수시장 동향 지수 확인 및 가공
car_kor.info()
```

```python
# [5-2] car_kor의 컬럼 이름 변경 및 필요한 행만 추출
car_kor2=car_kor.rename(columns={'Unnamed: 0':'목록'})
car_kor2=car_kor2.iloc[:2]
car_kor2.drop('2013',axis=1,inplace=True)
car_kor2
```

```python
# [6-1] 자동차 등록 대수 자료 확인 및 가공
car_update.info()
```

```python
# [6-2] car_updata의 필요한 자료만 추출
car_update2=car_update.rename(columns={'Unnamed: 0':'목록'})
car_update2
```

```python
# [6-3] DF 합성 ==> result
result=pd.concat([cpi3,car_kor2,car_update2],ignore_index=True)
```

소비자물가상승률과 내수시장 규모를 비교하고자 함

```python
# [7-1] result DF 해석
# 소비자 물가 상승률중 가장 큰 값, 중간값, 가장 작은 값일때의 내수시장 비교
result
```

```python
# [7-3] 소비자 물가 상승률 중 max, meaian, min 추출
print(f'소비자 물가 상승률 최댓값: :{result.iloc[1][:][1:].max()}')
print(f'소비자 물가 상승률 중앙값:{result.iloc[1][:][1:].median()}')
print(f'소비자 물가 상승률 최솟값 :{result.iloc[1][:][1:].min()}')
```

```python
# [7-2] 소비자물가상승률이 최댓값, 중앙값, 최솟값일 때의 데이터 출력
result[['목록','2014','2019','2022']]
```

소비자물가상승률과 내수 시장의 연관성이 없어 보여 우상향 하는 소비자물가 총지수와 내수 시장을 비교해보고자 함

```python
# [7-3] 내수가 최대,중앙,최소일때 소비자물가 총지수 데이터 비교
result[['목록','2020','2017','2014']]
```

1. 소비자물가상승률과 국산 차량의 내수시장은 연관성이 없어보임
2. 우상향 하는 소비자물가 총지수를 기반으로 내수시장을 비교해보니 소비자물가 총지수와 반비례하는 내수시장의 규모가 보임
3. 다만 2020년에 내수시장이 급격히 증가한 이유는 국가가 코로나19로 인한 소비 동결을 방지하기 위해 시행한 양적완화, 초저금리 정책의 영향으로 보임
4. 2020년 이후 내수시장의 감소를 통해 확인할 수 있음
5. 결론적으로 소비자물가와 국산 차량의 내수시장 규모는 반비례 하는 형태로 보임
6. 아쉬운점은 현재는 소비자물가와의 관계를 분석했을뿐 기업의 가격과 생산성의 정책 방향성을 고려하지 않은 결과물임
7. 기업의 정책 방향성을 추가 하고 차량 종별 가격 데이터를 추가 하여 금액적인 부분을 통해 단순 차량의 생산 대수가 아닌 내수시장의 활성화 정도를 파악하면 좋을듯함
