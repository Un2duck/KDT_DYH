```python
# 모듈로딩
import pandas as pd
import os
from collections import Counter
```

```python
# 데이터 파일
TRAIN_DATA = '../Data/lan_train/'
TEST_DATA = '../Data/lan_test/'

TRAIN_CSV = '../Data/lang_train.csv'
TEST_CSV = '../Data/lang_test.csv'
```

```python
# 파일 읽기 => open() -> read(), readline(), readlines(), ... ,  -> close()
#          == with open() as 구문

if not os.path.exists(TRAIN_CSV):
    print('TRAIN_CSV 파일이 없습니다.')
else:
    with open(TRAIN_CSV, mode='r') as f:
        alldata=f.read()
        print(f'[file Data]: {alldata}')
```

```python
def make_dataset(dir_path, file_path):
    # 폴더 안에 파일을 가져오기
    train_files = os.listdir(dir_path)

    for file in train_files:
        print(file, dir_path+file, os.path.exists(dir_path+file))
        label=file[:2]+','
        print(f'[label] => {label}')

        with open(dir_path+file, mode='r') as f:
            alldata=f.read()
            # print(f'[file Data]: {f.read()}')
        
        # 대소문자 통일
        alldata=alldata.lower()
        
        # 알파벳 아닌 문자들 제거
        alldata=alldata.replace('\n', '')
        print(f'alldata 길이 => {len(alldata)}')

        for ch in alldata:
            # if not ord('a')<=ch<=ord('z'):
            if ord('a')>ord(ch) or ord('z')<ord(ch):
                alldata=alldata.replace(ch, '')

        print(f'alldata 길이 => {len(alldata)}')
        print(f'alldata => {alldata}')

        # a ~ z까지 개수 카운팅
        # data={}
        # for ch in alldata:
        #     if ch in data:
        #         data[ch]+=1
        #     else:
        #         data[ch]=1
        # print(data)

        Count_alldata = Counter(alldata)
        # cc=dict(Count_alldata)
        cc2=dict(sorted(Count_alldata.items()))
        print(cc2)
        print(cc2.values())

        with open(file_path, mode='a') as wf:
            # wf.write(label)
            # wf.write(','.join([str(value) for value in cc2.values()])+'\n')
            a_z=[]
            for _ in range(ord('a'), ord('z')+1):
                # print(chr(_), alldata.count(chr(_)))
                cnt = alldata.count(chr(_))
                a_z.append(str(cnt/len(alldata)))

            wf.write(label+',')
            wf.write(','.join(a_z)+'\n')
```

```python
make_dataset(TRAIN_DATA, 'samples.csv')
make_dataset(TEST_DATA, 'samples2.csv')
```
