```python
from urllib.request import Request
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

```python
temp_list=[]
for idx in range(1, 10):
    url = f'https://www.health.kr/researchInfo/disease_detail.asp?idx={idx}'
    urlrequest = Request(url, headers = {'User-Agent' : 'Mozilla/5.0'})

    html = urlopen(urlrequest)
    bs = BeautifulSoup(html, 'html.parser')
    table = bs.find_all('p')
    table2 = bs.find_all('td') 
    if table2[2].text.isalpha():
        if table2[3].text == "호흡기계 질환":
            for row in table[1:-3]:
                print(row.text)
                temp_list.append(row.text)
            print('\n\n\n')
        else:
            pass
    elif table[3].text.isalpha():
        if table2[2].text == "호흡기계 질환":
            for row in table[1:-3]:
                print(row.text)
                temp_list.append(row.text)
            print('\n\n\n')
        else:
            pass
```

```python
def crawling(classifier):
    temp_list=[]
    for idx in range(1, 335):
        url = f'https://www.health.kr/researchInfo/disease_detail.asp?idx={idx}'
        urlrequest = Request(url, headers = {'User-Agent' : 'Mozilla/5.0'})

        html = urlopen(urlrequest)
        bs = BeautifulSoup(html, 'html.parser')
        table = bs.find_all('p')
        table2 = bs.find_all('td') 

        if classifier in table2[2].text:
            print(table2[2].text)
            for row in table[1:-3]:
                # print(row.text)
                temp_list.append(row.text)
            # print('\n\n\n')

        elif classifier in table2[3].text:
            print(table2[3].text)
            for row in table[1:-3]:
                # print(row.text)
                temp_list.append(row.text)
            # print('\n\n\n')

        else:
            pass
    return temp_list
```
