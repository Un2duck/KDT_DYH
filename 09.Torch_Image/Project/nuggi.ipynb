```python
# 모듈 로딩

from rembg import remove
from PIL import Image
import os
```

```python
input_path = './image/apple/Apple__Healthy/FreshApple (1).jpg' # 배경 제거할 이미지 경로
output_path = './image/apple/Apple_nugied/FreshApple (1).jpg' # 저장할 이미지 경로

img = Image.open(input_path)
out = remove(img)

# JPEG는 알파 채널을 지원하지 않으므로 RGB로 변환
if out.mode == 'RGBA':
    out = out.convert('RGB')

out.save(output_path)
```

```python
def check_files(dir_path):
    # 폴더 안에 파일 가져오기
    file_list = []
    files = os.listdir(dir_path)
    for file in files:
        print(f'file => \n{file}')
        file_list.append(file)

        # 배경 제거할 이미지 경로
        input_path = f'./image/strawberry/{file}'
        # 저장할 이미지 경로
        output_path = f'./image/strawberry_remove/{file}' 

        img = Image.open(input_path)
        out = remove(img)

        # JPEG는 알파 채널을 지원하지 않으므로 RGB로 변환
        if out.mode == 'RGBA':
            out = out.convert('RGB')

        out.save(output_path)
```

```python
# 폴더 경로
STRAWBERRY_PATH = './image/strawberry/'
```

```python
check_files(STRAWBERRY_PATH)
```
