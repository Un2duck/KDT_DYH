```python
import pandas as pd
import numpy as np
import os
```

```python
# 경로 확인 함수
def checkPath(PATH):
    if os.path.exists(PATH):
       print('경로있음')
    else:
        print('경로없음')
```

```python
# 데이터 파일 및 경로 지정
FILE = 'savefile.xlsx'
PATH = './Data/'
checkPath(PATH)
```

```python
# 데이터 확인
dataDF = pd.read_excel(PATH+FILE)
```

```python
# 가격 정수화하기
price=[]
for i in range(len(dataDF['가격'])): 
    price.append(int(dataDF['가격'][i].replace(',','')))
dataDF['가격']=pd.Series(price)
```

```python
dataDF.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
product_list=dataDF['상품명'].unique()
```

```python
# 가격체크 함수 생성
check_total=[]
def check_price(num):
    checkDF = dataDF[dataDF['상품명'] == product_list[num]]
    math_exp = round(checkDF['가격'].mean())-500
    check_list=[]
    for check in checkDF['가격'] >= math_exp:
        if check:
            check_list.append('정상가')
        else:
            check_list.append('의심됨')
    check_total.extend(check_list)
    return check_list
```

```python
[check_price(i) for i in range(len(product_list))]
```

```python
dataDF['어뷰징여부']=check_total
```
